## Dimensión VC
La dimensión VC (del inglés Vapnik-Chervonenkis dimension) es una medida de la capacidad de los algoritmos de clasificación estadística, definida como la cardinalidad del mayor conjunto de puntos que el algoritmo puede separar. Es un concepto central en la teoría Vapnik-Chervonenkis, y fue originalmente definido por Vladimir Vapnik y Alexey Chervonenkis. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

Un modelo de clasificación f con algún vector de parámetros θ es llamado separador de un conjunto de puntos (x1,x2,…,xn) si, para todas las asignaciones de las etiquetas de esos puntos, existe una θ tal que el modelo f no tiene errores cuando evalúa ese conjunto de puntos. 

La dimensión VC del modelo f es el máximo h para el cual algún punto del conjunto de datos de cardinalidad h puede ser separado por f. 



```python
x = np.linspace(1,20,300)
f_dvc = [(x**i)*np.exp(-x) for i in [2,3,5,10,15,20]]
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(1,1,1)

ax.set_yscale('log')

for a,b in zip (f_dvc, [2,3,5,10,15,20]):
    plt.plot(x,a, label = "{} dimensiones VC".format(b))
plt.legend()
```
